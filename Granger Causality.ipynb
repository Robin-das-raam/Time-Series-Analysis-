```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from statsmodels.tsa.stattools import grangercausalitytests
```

```python
# build a simple time series with ar(1)

t1 = [0.1*np.random.normal()]

for _ in range(100):
    t1.append(0.5*t1[-1] + 0.1*np.random.normal())
```

```python
t1
```

```python
# build another simple time series that is granger caused by t1

t2 = [item + 0.1* np.random.normal() for item in t1]
t2
```

```python
#adjust t1 and t2
t1 = t1[3:]
t2 = t2[:-3]
```

```python
plt.figure(figsize = (10,4))
plt.plot(t1, color = "r")
plt.plot(t2, color = "b")

plt.legend(["T1", "T2"], fontsize = 15)
```
